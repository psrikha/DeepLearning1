Function 1 : sin(5*pi*x) / (5*pi*x)

```python
import numpy as np
import matplotlib.pyplot as plt
X = np.arange(-10,4*np.pi,0.1)
X = X*0.095
y = (np.sin(5*(np.pi*X)))/((5*(np.pi*X)))
plt.plot(X,y)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
X_t = torch.tensor(X,dtype=torch.float32)
y_t = torch.tensor(y,dtype=torch.float32)
X_t = X_t.view(X_t.shape[0],1)
y_t = y_t.view(y_t.shape[0],1)
```

Function1, Model 0

```python
class func_simulator0(nn.Module):
    def __init__(self):
        super(func_simulator0,self).__init__()
        self.l1 = nn.Linear(1,128)
        self.l2 = nn.Linear(128,1)
    def forward(self,x):
        out = F.relu(self.l1(x))
        out = self.l2(out)
        return out
```

```python
learning_rate,num_epochs    = 0.001,2000
model                       = func_simulator0()
loss                        = nn.MSELoss()
model0_losses = []
gradient = torch.optim.SGD(model.parameters(),lr=learning_rate)
# start training
for epoch in range(num_epochs):
    y_pred = model.forward(X_t)
    l = loss(y_pred,y_t)
    model0_losses.append(l)
    l.backward()
    gradient.step()
    gradient.zero_grad()
    if(epoch+1)%1000==0:
        print(f'epoch:{epoch+1},loss={l.item():.3f}')
print(f'losses:{model0_losses}')
    
```

```python
predicted = model(X_t).detach().numpy()
plt.plot(X_t,y_t)
plt.plot(X_t,predicted,'b')
plt.show()
```
